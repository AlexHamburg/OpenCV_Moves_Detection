```python
from bokeh.plotting import figure, output_file, show
import pandas as pd
```

```python
csv = pd.read_csv("adbe.csv", parse_dates = ["Date"])
x = csv.iloc[:,0].values
y = csv.iloc[:,1].values
```

```python
output_file("line_2.html")
plot = figure(plot_width = 500, plot_height = 200, x_axis_type = "datetime", 
              tools = "pan,reset", logo = None, title = "Title")

plot.title.text_color = "Red"
plot.title.text_font = "times"
plot.title.text_font_style = "italic"
plot.sizing_mode = "scale_both"
plot.xaxis.minor_tick_line_color = "Yellow"

plot.xaxis.axis_label = "Time"
plot.yaxis.axis_label = "Values"
plot.line(x,y, line_width = 2, color = "red", alpha = 0.5)
#plot.circle(x,y, size = [i*2 for i in [1, 2, 3, 4, 5]], color = "navy", alpha = 0.5)
```

```python
show(plot)
```
